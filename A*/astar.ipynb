```python
import math
from typing import Optional

grafo_no = { 'S':{'A':3},
               'A':{'B':1, 'C':2,'D': 3},
               'B':{'E':4},
               'C':{'E':5},
               'D':{'E':1},
               'E':{}
}

heuristica = {
    'S':5,
    'A':4,
    'B':3,
    'C':4,
    'D':1,
    'E':0
}

def montar_caminho(pai: dict[str, str], inicio: str, fim: str) -> list[str]:
    caminho = [fim]
    atual = fim
    while atual != inicio:
        atual = pai[atual]
        caminho.append(atual)
    caminho.reverse()
    return caminho

def busca_melhor_primeiro(no_inicial: str,no_objetivo: str):
    no_atual = no_inicial
    fronteira: list[tuple[str,int]] = []
    alcançados = set()
    g_x = {no_inicial:0}

    pai: dict[str, str] = {}  # <-- ADICIONADO

    fronteira.append((no_atual, heuristica[no_atual]))
    while len(fronteira) > 0:
        if no_atual == no_objetivo:
            return montar_caminho(pai, no_inicial, no_objetivo)  # <-- ALTERADO

        alcançados.add(no_atual)

        g_x = expandir(no_atual, g_x, pai)  # <-- ALTERADO
        for vizinho in grafo_no[no_atual].keys():
            if vizinho not in alcançados:
                fronteira.append((vizinho, heuristica[vizinho] + g_x.get(vizinho, math.inf)))

        fronteira.sort(key=lambda x: x[1])
        no_atual, _ = fronteira.pop(0)

    return "Falha"


def expandir(no_atual: str, g_x: dict, pai: dict[str, str]) -> dict:  # <-- ALTERADO
    for vizinho, custo in grafo_no[no_atual].items():
        novo_g = custo + g_x[no_atual]
        if g_x.get(vizinho) is None or g_x.get(vizinho) > novo_g:
            g_x[vizinho] = novo_g
            pai[vizinho] = no_atual  # <-- ADICIONADO (guarda o caminho)
    return g_x

resultado = busca_melhor_primeiro('S','E')
print(resultado)
```
